```python
#Temperature function
def f(x,y):
    return 2*x*y + 2*x - x**2 -2*(y**2) + 72

#Analytical Solution
def sol(x,y):
    return 0.5*(x**2)*(y**2) + (x**2)*y -((x**3)/3)*y - (2*(y**3)/3)*x + 72*x*y

exact = sol(8,6)

#Trapezoidal Method
def trap(a1, a2, b1, b2, n):
    z = [[None for i in range(1000)]
           for j in range(1000)]
    ay = [None]*1000
    h1 = (b1 - a1)/n
    h2 = (b2 - a2)/n
    
    for j in range(0,n+1):
        for i in range(0,n+1):
            z[i][j] = f(a1 + i*h1, a2 + j*h2)
    for j in range(0,n+1):
        ay[j] = 0
        for i in range(0,n+1):
            if i == 0 or i == n:
                ay[j] += z[i][j]
            else:
                ay[j] += 2*z[i][j]
        ay[j] *= h1/2
    #print(ay)
    a = 0
    for j in range(0,n+1):
        if j == 0 or j == n:
            a += ay[j]
        else:
            a += 2*ay[j]
    a *= h2/2
    return a      

a1 = 0
a2 = 0
b1 = 8
b2 = 6
n = 2
t = trap(a1, a2, b1, b2, n)


#Simpson's 1/3rd Rule
def simp13(a1, a2, b1, b2, n):
    z = [[None for i in range(1000)]
           for j in range(1000)]
    ay = [None]*1000
    h1 = (b1 - a1)/n
    h2 = (b2 - a2)/n
    
    for i in range(0,n+1):
        for j in range(0,n+1):
            z[i][j] = f(a1 + i*h1, a2 + j*h2)
    for j in range(0,n+1):
        ay[j] = 0
        for i in range(0,n+1):
            if i == 0 or i == n:
                ay[j] += z[i][j]
            elif i % 2 == 0:
                ay[j] += 2*z[i][j]
            else:
                ay[j] += 4*z[i][j]
        ay[j] *= h1/3
    #print(ay)
    a = 0
    for j in range(0,n+1):
        if j == 0 or j == n:
            a += ay[j]
        elif j % 2 == 0:
            a += 2*ay[j]
        else:
            a += 4*ay[j]
    a *= h2/3
    return a  

s13 = simp13(a1, a2, b1, b2, n)


#Simpson's 3/8th Rule

def simp38(a1, a2, b1, b2):
    z = [[None for i in range(100)]
           for j in range(100)]
    ay = [None]*100
    h1 = (b1 - a1)/3
    h2 = (b2 - a2)/3
    
    for i in range(0,4):
        for j in range(0,4):
            z[i][j] = f(a1 + i*h1, a2 + j*h2)
    for j in range(0,4):
        ay[j] = 0
        for i in range(0,4):
            if i == 0 or i == 3:
                ay[j] += z[i][j]
            else:
                ay[j] += 3*z[i][j]
        ay[j] *= h1*(3/8)
    #print(ay)
    a = 0
    for j in range(0,4):
        if j == 0 or j == 3:
            a += ay[j]
        else:
            a += 3*ay[j]
    a *= h2*(3/8)
    return a 


def err(a, s):
    e = (s-a)/s * 100
    return e
        
s38 = simp38(a1, a2, b1, b2)

print("The average temperature through Trapezoidal method is =" , t/48, "The relative error is", err(t, sol(8,6)),"%")
print("The average temperature through Simpson's 1/3rd method is =" , s13/48, "The relative error is", err(s13, sol(8,6)),"%")
print("The average temperature through Simpson's 3/8th method is =" , s38/48 , "The relative error is", err(s38, sol(8,6)),"%")
print("The average temperature is exactly =" , exact/48)



```
