```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
def approx_pi(m):
    
    Ns = m
    Nc = 0

    Z = np.random.rand(m)  
    
    for i in range(int(m-1)):

        if ((Z[i]**2 + Z[i+1]**2)**0.5 < 1 ):
            Nc += 1

    pi_estimate = 4* Nc / Ns

    return pi_estimate
```

```python
def approx_pi_lcg(n, a, c):

    Ns = n
    Nc = 0
    m = n
    x = np.zeros(n)
    
    x[0] = 67
    
    for i in range(1, n):

        x[i] = (((a * x[i-1]) + c) % (m) )
        
    x /= n
    y = np.zeros(n)
    
    y[0] = 733
    a = 3079
    c = 4096
    for i in range(1, n):

        y[i] = (((a * y[i-1]) + c) % (m) ) 
        
    y /= n
    for i in range(int(n)):

        if ((x[i]**2 + y[i]**2)**0.5 < 1 ):
            Nc += 1

    pi_estimate = 4 * Nc / Ns

    return pi_estimate
```

```python
def main(n):
    x = np.arange(1, n+1, 1)
    v_pi = np.zeros(n)
    sum = 0

    for i in range(n):
        v_pi[i] = approx_pi(x[i])
        sum += v_pi[i] 
    
    print(sum/n)    
    
    plt.plot(x, v_pi)
    
    exact = np.ones(n)
    exact *= pi
          
    plt.plot(x, exact)

    plt.show()    
    sum = 0
    for i in range(n):
    
        v_pi[i] = approx_pi_lcg(x[i], 887, 701)
        sum += v_pi[i] 

    print(sum/n)
    plt.plot(x, v_pi)
    plt.plot(x, exact)

    plt.show()
```
